```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("C:\\Users\\User\\Desktop\\AI and ML\\bigdata1.csv")
print(df)
plt.bar(df['Period'],df['Region'])
plt.xlabel('x-axis label')
plt.ylabel('y-axis label')
plt.title('barplot')
plt.show()
plt.stem(df['Count'],df['Region'])
plt.xlabel('x-axis label')
plt.ylabel('y-axis label')
plt.title('stemplot')
plt.show()
plt.stackplot(df['Count'],df['Region'])
plt.xlabel('x-axis label')
plt.ylabel('y-axis label')
plt.title('stackplot')
plt.show()
plt.scatter(df['Period'],df['Count'])
plt.xlabel('x-axis label')
plt.ylabel('y-axis label')
plt.title('scatterplot')
plt.show()
plt.hist(df['Period'],weights=df['Count'])
plt.xlabel('Period')
plt.ylabel('Count')
plt.title('histogram')
plt.show()

```
